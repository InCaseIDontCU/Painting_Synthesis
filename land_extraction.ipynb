```python
from PIL import Image
import tensorflow as tf
from tensorflow.keras.applications.vgg16 import VGG16
import numpy as np
import os
```

```python
# Load images in a directory
def load_image():
    images = list()
    dir = "landscape/sub"
    for filename in os.listdir(dir):
        paint = Image.open(dir + "/" + filename)
        paint.show()
        normsize = 224
        newpaint = paint.resize((normsize, normsize))
        newpaint = newpaint.convert("RGB")
        newpaint.show()
        pixel = np.asarray(newpaint)
        images.append(pixel)
    return np.asarray(images)
```

```python
# Landscape feature extraction
def land_extra(pixel_list):
    model = VGG16(weights='imagenet')
    model = tf.keras.Model(inputs=model.input, outputs=model.get_layer("block5_pool").output)
    model.trainable = False
    result = []
    
    pred = model.predict(pixel_list)
    result.append(pred)
    return np.squeeze(np.array(result), axis=0)
```

```python
pixels = load_image()
landscape = land_extra(pixels)
```

```python
np.savez_compressed("landscape/landscape1.npz",landscape)
```
